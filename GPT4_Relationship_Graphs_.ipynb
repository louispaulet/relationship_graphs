```python

import json
import networkx as nx
import matplotlib.pyplot as plt

# JSON data as a large string
json_data = '''
{
  "nodes": [
    { "id": "michael", "label": "Michael Scott" },
    { "id": "jim", "label": "Jim Halpert" },
    { "id": "pam", "label": "Pam Beesly" },
    { "id": "dwight", "label": "Dwight Schrute" },
    { "id": "angela", "label": "Angela Martin" },
    { "id": "oscar", "label": "Oscar Martinez" },
    { "id": "kevin", "label": "Kevin Malone" },
    { "id": "stanley", "label": "Stanley Hudson" },
    { "id": "phyllis", "label": "Phyllis Vance" },
    { "id": "meredith", "label": "Meredith Palmer" }
  ],
  "links": [
    { "source": "michael", "target": "jim", "label": "boss" },
    { "source": "michael", "target": "pam", "label": "boss" },
    { "source": "michael", "target": "dwight", "label": "boss" },
    { "source": "dwight", "target": "jim", "label": "colleague" },
    { "source": "dwight", "target": "angela", "label": "romantic partner" },
    { "source": "angela", "target": "oscar", "label": "colleague" },
    { "source": "kevin", "target": "oscar", "label": "colleague" },
    { "source": "stanley", "target": "phyllis", "label": "colleague" },
    { "source": "meredith", "target": "kevin", "label": "colleague" }
  ]
}
'''

# Load JSON data from the string
data = json.loads(json_data)

# Create an empty graph
G = nx.Graph()

# Add nodes to the graph
for node in data['nodes']:
    G.add_node(node['id'], label=node['label'])

# Add edges to the graph
for link in data['links']:
    G.add_edge(link['source'], link['target'], label=link['label'])

# Generate a layout for the graph
pos = nx.spring_layout(G)

# Draw nodes with labels
labels = nx.get_node_attributes(G, 'label')
nx.draw_networkx_nodes(G, pos, node_size=200, alpha=0.8)
nx.draw_networkx_labels(G, pos, labels, font_size=16)

# Draw edges with labels
edge_labels = nx.get_edge_attributes(G, 'label')
nx.draw_networkx_edges(G, pos, width=1.0, alpha=0.5)
nx.draw_networkx_edge_labels(G, pos, edge_labels, font_size=14)

# Display the graph
plt.axis('off')
plt.show()
```

## define general function (change directory if needed here also)

```python
import json
import networkx as nx
import matplotlib.pyplot as plt
import time
from datetime import datetime


save_path = "/content/drive/MyDrive/AI/graphs/graph_images/"


def display_json_data(json_data, sizeX=15, sizeY=15):
    # Load JSON data from the string
    data = json.loads(json_data)

    # Create an empty graph
    G = nx.Graph()

    # Add nodes to the graph
    for node in data['nodes']:
        G.add_node(node['id'], label=node['label'])

    # Add edges to the graph
    for link in data['links']:
        G.add_edge(link['source'], link['target'], label=link['label'])

    # Generate a layout for the graph
    pos = nx.spring_layout(G)

    # Increase the figure size
    plt.figure(figsize=(sizeX, sizeY))

    # Draw nodes with labels
    labels = nx.get_node_attributes(G, 'label')
    nx.draw_networkx_nodes(G, pos, node_size=2000, alpha=0.8)
    nx.draw_networkx_labels(G, pos, labels, font_size=10)

    # Draw edges with labels and increase edge length
    edge_labels = nx.get_edge_attributes(G, 'label')
    nx.draw_networkx_edges(G, pos, width=1.0, alpha=0.5, edge_color='black')
    nx.draw_networkx_edge_labels(G, pos, edge_labels, font_size=8, label_pos=0.5, rotate=False)

    # Save the graph image with a filename prefixed with a timestamp (including milliseconds)
    # timestamp = time.strftime("%Y%m%d%H%M%S%f")
    timestamp = datetime.now().strftime("%H:%M:%S.%f")
    filename = f"{save_path}graph_{timestamp}.png"
    plt.savefig(filename)

    # Display the graph
    plt.axis('off')
    plt.show()

    # Close the plot to release memory
    plt.close()
```
